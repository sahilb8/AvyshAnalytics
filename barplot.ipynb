```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
compDataset = pd.read_csv('Data Population - Main2.csv')
```

```python
compDataset
```

```python
sellerDetails=pd.read_csv('Data Population - Sellers.csv')
```

```python
sellerDetails
```

```python
#unique seller id in a list
sellerID=sellerDetails['Sellers ID'].tolist()
```

```python
sellerID
```

```python
#all the seller ids in the main transaction dataframe
mainSellerID=compDataset['Sellers_ID'].tolist()
```

```python
mainSellerID
```

```python
#number of transaction for each sellerID
numOfTransactions=[]
for i in sellerID:
    c=0
    for j in range(0,1018):
        if(i==mainSellerID[j]):
            c+=1
    numOfTransactions.append(c)
```

```python
numOfTransactions
```

```python
#Code to remove the spaces in the column names and replace space with '_'
compDataset.columns = [c.replace(' ', '_') for c in compDataset.columns]
```

```python
#list of all the values
value=compDataset['Value'].tolist()
```

```python
value
```

```python
len(mainSellerID)
```

```python
val_perSeller=[]
for i in sellerID:
    val_sum=0
    for j in range(0,1018):
        if(i==mainSellerID[j]):
            val_sum+=value[j]
    val_perSeller.append(val_sum)
    
```

```python
val_perSeller
```

```python
#dictionary for seller ID and respective value
sellerID_val= dict(zip(sellerID, val_perSeller))
```

```python
sellerID_val
```

```python
#average value per seller
avg_val=[]   
for i in range(0,140):
    a=val_perSeller[i]
    b=numOfTransactions[i]
    c=a/b
    avg_val.append(c)
```

```python
avg_val
```

```python
#dictionary for seller ID and respective average value of transaction
sellerID_avgVal= dict(zip(sellerID, avg_val))
```

```python
sellerID_avgVal
```

```python
#main bar plot of number of transactions per seller, further then devided the graph in 3 parts
plt.bar(sellerID,numOfTransactions,alpha=0.9)
plt.xlabel('Seller_ID')
plt.ylabel('Num Of Transactions')
plt.title('Num of transactions per seller plot')
#plt.show()
plt.xticks(sellerID, [str(i) for i in sellerID], rotation=90)

#set parameters for tick labels
plt.tick_params(axis='x', which='major', labelsize=3)

plt.tight_layout()
```
