```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('/home/sahil/Documents/AvyshCodeAndData/Data Population - Main2.csv')
```

```python
df.head()
```

```python
#Code to remove the spaces in the column names and replace space with '_'
df.columns = [c.replace(' ', '_') for c in df.columns]
```

```python
#Looking only at the data of france
basket = (df.groupby(['Sellers_ID', 'SKU'])['Qty']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Sellers_ID'))
```

```python
basket
```

```python
# Show a subset of columns
basket.iloc[:,[0,1,2,3,4,5,6, 7]].head()
```

```python
# Convert the units to 1 hot encoded values
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_sets = basket.applymap(encode_units)
#applymap is apply a function to the dataframe element wise
```

```python
basket_sets.head()
```

```python
# Build up the frequent items
frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets.head()
```

```python
# Create the rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.4) ]
```
